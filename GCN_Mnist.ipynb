```python
from tqdm import tqdm
import networkx as nx
import numpy as np
import torch
from sklearn.preprocessing import StandardScaler
import pandas as pd
from torch_geometric.data import InMemoryDataset, Data
from sklearn.model_selection import train_test_split
import torch_geometric.transforms as T
import torch.nn as nn
import torch.nn.functional as F
from torch_geometric.nn import GCNConv
from torchvision import datasets, transforms
from torchmetrics.functional import pairwise_cosine_similarity
from sklearn.metrics.pairwise import cosine_similarity as cos_similarity
import time
import torch.optim as optim
from sklearn.metrics import ConfusionMatrixDisplay
from torch_geometric.utils.convert import to_networkx
import matplotlib.pyplot as plt 
```

```python
#returns torch tensors
def get_data(dataset):
    X = []
    y = []
    for i in range(dataset.__len__()):
        img, label = dataset.__getitem__(i)
        X.append(img)
        y.append(label)

    X = torch.cat(X, 0)
    X = X.reshape(X.shape[0],-1)
    y = torch.tensor(y)
    return X , y

def extract_data(train_data, test_data):
    X_train, y_train = get_data(train_data)
    X_test, y_test = get_data(test_data)
    X = torch.cat([X_train, X_test],0)
    y = torch.cat([y_train, y_test],0)
    return X, y
```

```python
def get_edges(X, k=10):
    #k = 10
    edges = torch.zeros(2, X.shape[0]*k).long()

    start_pointer=0
    for i in tqdm(range(X.shape[0])):
        sim = pairwise_cosine_similarity(X[i].unsqueeze(0),X)
        sim = sim[0]
        sim[i] = 0
        neighbour_i = torch.topk(sim, k).indices

        edges[0][start_pointer:start_pointer+k] = i
        edges[1][start_pointer:start_pointer+k] = neighbour_i
        start_pointer+=k
        
    return edges
```

```python
# custom dataset
class MnistDataset(InMemoryDataset):
    def __init__(self, embeddings, labels, edges, transform=None):
        super(MnistDataset, self).__init__('.', transform, None, None)
        self.X = embeddings
        self.y = labels
        self.edges = edges
        
        data = Data(edge_index=self.edges)
        data.num_nodes = self.X.shape[0]
        
        # embedding 
        data.x = self.X
        #labels
        data.y = self.y

        data.num_classes = 10

        len_train_set = 60000
        
        n_nodes = X.shape[0]
        
        # create train and test masks for data
        train_mask = torch.zeros(n_nodes, dtype=torch.bool)
        test_mask = torch.zeros(n_nodes, dtype=torch.bool)
        train_mask[:len_train_set] = True
        test_mask[len_train_set:] = True
        data['train_mask'] = train_mask
        data['test_mask'] = test_mask

        self.data, self.slices = self.collate([data])
        
    def _download(self):
        return

    def _process(self):
        return

    def __repr__(self):
        return '{}()'.format(self.__class__.__name__)
```

```python
# GCN model with 4 layers 
class GCN(torch.nn.Module):
    def __init__(self):
        super(GCN, self).__init__()
        self.conv1 = GCNConv(data.num_features, 300)
        self.RELU1 = self.RELU(300)
        
        self.conv2 = GCNConv(300, 200)
        self.RELU2 = self.RELU(200)
        
        self.conv3 = GCNConv(200, 64)
        self.RELU3 = self.RELU(64)
        
        self.conv4 = GCNConv(64, int(data.num_classes))
        self.batchnorm4 = nn.BatchNorm1d(data.num_classes)
        #self.RELU4 = self.RELU(int(data.num_classes))
        
    def RELU(self,out_channels):
        return nn.Sequential(nn.BatchNorm1d(out_channels),
                             nn.LeakyReLU(0.1))
    
    def forward(self):
        x, edge_index = data.x, data.edge_index
        x = self.RELU1(self.conv1(x, edge_index))
        x = self.RELU2(self.conv2(x, edge_index))
        x = self.RELU3(self.conv3(x, edge_index))
        x = self.batchnorm4(self.conv4(x, edge_index))
        return x
```

```python
def cross_entropy_loss(predictions, target):
    loss = nn.CrossEntropyLoss()
    return loss(predictions, target)
```

```python
def train_fn_(data, model, optimizer, epoch):
    model.train()
    optimizer.zero_grad()
    loss = cross_entropy_loss(model()[data.train_mask], data.y[data.train_mask])
    loss.backward()
    optimizer.step() 
    return loss.item()
```

```python
@torch.no_grad()
def eval_model(data, model, optimizer):
    model.eval()
    logits = model()
    mask1 = data['train_mask']
    pred1 = logits[mask1].max(1)[1]
    train_acc = pred1.eq(data.y[mask1]).sum().item() / mask1.sum().item()
    mask = data['test_mask']
    pred = logits[mask].max(1)[1]
    test_loss = cross_entropy_loss(logits[data.test_mask], data.y[data.test_mask])
    test_acc = pred.eq(data.y[mask]).sum().item() / mask.sum().item()
    return train_acc, test_acc, test_loss
```

```python
def save_checkpoint(state, filename="my_checkpoint.pth"):
    print("=> Saving checkpoint")
    torch.save(state, filename) 
```

```python
def load_checkpoint(checkpoint, model, optimizer):
    print("=> Loading checkpoint")
    model.load_state_dict(checkpoint["state_dict"])
    optimizer.load_state_dict(checkpoint["optimizer"])
```

```python
#Hyperparameters
#LEARNING_RATE = 2e-5
LEARNING_RATE = 1e-1
#WEIGHT_DECAY = 1e-4
EPOCHS = 150
LOAD_MODEL = False
LOAD_MODEL_FILE = "<Give the file path where you want to store model + /model_gcn.pth>" 
DEVICE = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

```python
#Load data

#transforms
transform = transforms.Compose([transforms.ToTensor(),
                              #transforms.Normalize((0.5,), (0.5,)),
                              ])

dir_path = "<Give the directory path where you want to store dataset>"
train_data = datasets.MNIST(dir_path, download=True, train=True, transform=transform)
test_data = datasets.MNIST(dir_path, download=True, train=False, transform=transform)
X, y = extract_data(train_data, test_data)

#Load edges from csv
filepath = "<Give path of edges.csv>"
edges_df = pd.read_csv(filepath)
#edges_df.drop(columns=['Unnamed: 0'], inplace=True)
edges = np.array(edges_df)
edges = torch.tensor(edges).long()
print(edges.shape)
dataset = MnistDataset(X, y, edges)
data = dataset[0]
print(data)
```

```python
#Let us visualize a subset of large graph

edges1 = get_edges(X[:20], k=10)
dataset1 = MnistDataset(X[:20], y[:20], edges1)
data1 = dataset1[0]
data1.train_mask =None
data1.test_mask =None
print(data1)
vis = to_networkx(data1)
node_labels = data1.y.numpy()
plt.figure(1,figsize=(10,7))

nx.draw(vis, node_color =node_labels,cmap=plt.get_cmap('Set3'), node_size=200, linewidths=3)
print("Here, each node is a image of MNIST dataset")
```

```python
#main

#Load model
model = GCN().to(DEVICE)

optimizer = optim.Adam(
        model.parameters(), lr=LEARNING_RATE
    )

if LOAD_MODEL:
        load_checkpoint(torch.load(LOAD_MODEL_FILE), model, optimizer)


train_loss_epochs = []
train_acc_epochs = []
test_loss_epochs = []
test_acc_epochs = []

for epoch in tqdm(range(EPOCHS),bar_format='{l_bar}{bar:10}{r_bar}{bar:-10b}'):
    
    train_loss = train_fn_(data, model, optimizer, epoch)
    train_acc, test_acc, test_loss = eval_model(data, model, optimizer)
    
    train_loss_epochs.append(train_loss)
    train_acc_epochs.append(train_acc)
    test_loss_epochs.append(test_loss)
    test_acc_epochs.append(test_acc)
    
checkpoint = {
            "state_dict": model.state_dict(), 
            "optimizer": optimizer.state_dict(), 
             } 
             
save_checkpoint(checkpoint, filename=LOAD_MODEL_FILE)  
```

```python
#Plotting training and testing curves
plt.figure(figsize=(10, 5))
plt.subplot(1,2,1)
plt.plot(train_loss_epochs)
plt.plot(train_acc_epochs)
plt.legend(['Train loss','Train Accuracy'])
plt.xlabel('Loss / Accuracy')
plt.ylabel('Epochs')
plt.title('Training curves')

plt.subplot(1,2,2)
plt.plot(test_loss_epochs)
plt.plot(test_acc_epochs)
plt.legend(['Test loss','Test Accuracy'])
plt.xlabel('Loss / Accuracy')
plt.ylabel('Epochs')
plt.title('Testing curves')
```
